```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Siddhant\Downloads\data\data\Bank_churn_modelling.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df.CreditScore[df.Exited==0])
sns.distplot(df.CreditScore[df.Exited==1])
plt.legend(['0','1'])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df.Age[df.Exited==0])
sns.distplot(df.Age[df.Exited==1])
plt.legend(['0','1'])
plt.show()##The below graph shows that majority of people leaving are from old age group ,young age  has less drop out rate
```

```python
plt.figure(figsize=(6,3))
sns.countplot(df.Geography)
plt.show()
sns.countplot(df.Geography[df.Exited==1])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(df.Gender[df.Exited==1])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(df.HasCrCard[df.Exited==0])
plt.show()
plt.figure(figsize=(12,5))
sns.countplot(df.HasCrCard[df.Exited==1])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.pointplot(x='Geography',y='Age',hue='Exited',data=df)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='CreditScore',y='Age',hue='Exited',data=df)
plt.show()
```
